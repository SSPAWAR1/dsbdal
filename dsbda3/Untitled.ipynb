```python
import pandas as pd
df = pd.read_csv("iris2.csv")

```

```python
# Set display precision
pd.set_option('display.precision', 2)

# Set display width
pd.set_option('display.width', 100)

# Group by a categorical variable and calculate summary statistics
summary_stats = df.groupby('species').agg({
    'sepal_length': ['mean', 'median', 'min', 'max', 'std'],
    'sepal_width': ['mean', 'median', 'min', 'max', 'std'],
    'petal_length': ['mean', 'median', 'min', 'max', 'std'],
    'petal_width': ['mean', 'median', 'min', 'max', 'std']
})

print(summary_stats)
```
